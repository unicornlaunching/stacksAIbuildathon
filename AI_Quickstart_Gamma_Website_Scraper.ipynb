```python
import requests
from bs4 import BeautifulSoup

# URL of the page containing all the links
main_url = "https://stacks.gamma.io/learn"

# Function to scrape text from a page given its URL
def scrape_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Extract the content from the <div class="mb-10">
        content = soup.find('div', class_='mb-10')
        if content:
            return content.get_text(strip=True)
        else:
            return "Content not found."
    else:
        return f"Failed to retrieve the page. Status code: {response.status_code}"

# Fetch the main page to get all the links
response = requests.get(main_url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the links on the page
    links = soup.find_all('a', href=True)

    # Filter links that contain "learn/" in the URL
    learn_links = [link['href'] for link in links if '/learn/' in link['href']]

    # Loop through each link, visit the page, and scrape the content
    for link in learn_links:
        # Full URL to the linked page
        full_url = f"https://stacks.gamma.io{link}"
        print(f"Scraping: {full_url}")
        page_content = scrape_page(full_url)
        print(page_content)
        print("\n" + "="*80 + "\n")  # Separator between pages
else:
    print(f"Failed to retrieve the main page. Status code: {response.status_code}")
```

```python
# Open a file to save the results
with open('scraped_content.txt', 'w', encoding='utf-8') as f:
    for link in learn_links:
        full_url = f"https://stacks.gamma.io{link}"
        print(f"Scraping: {full_url}")
        page_content = scrape_page(full_url)
        f.write(f"URL: {full_url}\n")
        f.write(page_content + "\n")
        f.write("\n" + "="*80 + "\n")
```

```python
import os
import requests
from bs4 import BeautifulSoup
from zipfile import ZipFile

# URL of the page containing all the links
main_url = "https://stacks.gamma.io/learn"

# Function to scrape headers, paragraphs, and title from a page
def scrape_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract the title of the page
        title_tag = soup.find('title')
        page_title = title_tag.get_text(strip=True) if title_tag else "untitled"

        # Extract all headers and paragraphs
        headers_and_paragraphs = []

        # Extract headers (h1, h2, h3, etc.)
        for header in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
            headers_and_paragraphs.append(header.get_text(strip=True))

        # Extract paragraphs
        for paragraph in soup.find_all('p'):
            headers_and_paragraphs.append(paragraph.get_text(strip=True))

        # Join all the text
        page_content = "\n".join(headers_and_paragraphs)

        return page_title, page_content
    else:
        return None, None

# Function to create a valid file name from the title
def create_file_name_from_title(title):
    # Replace spaces and special characters to make a valid file name
    file_name = "".join(x if x.isalnum() or x in " _-" else "_" for x in title)
    # Add .txt extension
    return f"{file_name}.txt"

# Fetch the main page to get all the links
response = requests.get(main_url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the links on the page
    links = soup.find_all('a', href=True)

    # Filter links that contain "learn/" in the URL
    learn_links = [link['href'] for link in links if '/learn/' in link['href']]

    # Create a directory to store the text files
    if not os.path.exists('scraped_articles'):
        os.makedirs('scraped_articles')

    # List to hold all file paths for zipping later
    file_paths = []

    # Loop through each link, visit the page, scrape the content, and save to a .txt file
    for link in learn_links:
        # Full URL to the linked page
        full_url = f"https://stacks.gamma.io{link}"
        print(f"Scraping: {full_url}")

        # Scrape the page content and title
        page_title, page_content = scrape_page(full_url)

        if page_content:
            # Create a valid file name based on the page title
            file_name = create_file_name_from_title(page_title)

            # Path for saving the file
            file_path = os.path.join('scraped_articles', file_name)

            # Save the content to a .txt file
            with open(file_path, 'w', encoding='utf-8') as f:
                f.write(page_content)

            # Add the file path to the list for zipping
            file_paths.append(file_path)
            print(f"Saved: {file_path}")
        else:
            print(f"Failed to retrieve content from: {full_url}")

    # Create a zip file to compress all the .txt files
    zip_file_name = "scraped_articles.zip"
    with ZipFile(zip_file_name, 'w') as zipf:
        for file_path in file_paths:
            zipf.write(file_path, os.path.basename(file_path))

    print(f"All articles have been zipped into {zip_file_name}")
else:
    print(f"Failed to retrieve the main page. Status code: {response.status_code}")
```
