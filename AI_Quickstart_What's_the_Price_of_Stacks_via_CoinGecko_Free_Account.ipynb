# 1. Let's dial the CoinGecko API Hotline
#### You'll Need an API Key aka Password from CoinGecko to use their API Hotline. You can Sign up and get an API key here: https://www.coingecko.com/
#### I'm using the docs provided by CoinGecko here: https://docs.coingecko.com/v3.0.1/reference/introduction

```python
# Import the requests library which helps us make HTTP requests
import requests

# The API endpoint we're trying to talk to
url = "https://api.coingecko.com/api/v3/ping"

# This is the header where we include our API key (replace 'YOUR_API_KEY' with your actual API key)
headers = {
    'x-cg-demo-api-key': 'INSERT_API_KEY_HERE'  # This is like showing an ID badge to get access
}

# We are sending a GET request (asking the API for information)
response = requests.get(url, headers=headers)

# Check if the request worked (status code 200 means success)
if response.status_code == 200:
    # If it worked, let's print the response data
    print(response.json())  # This shows us the data the API sent back, in this case, it 'pings' back
else:
    # If it didn't work, let's print the error code and message
    print(f"Error: {response.status_code}, {response.reason}")
```

### Yay! We got a response that says, '(V3) To The Moon!

# 2. Let's dial the CoinGecko Hotline and Get Bitcoin information

```python
import requests

url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd&include_market_cap=true&include_24hr_vol=true&include_24hr_change=true&include_last_updated_at=true&precision=full"

headers = {
    "accept": "application/json",
    "x-cg-demo-api-key": "INSERT_API_KEY_HERE\t"
}

response = requests.get(url, headers=headers)

print(response.text)
```

### Yay! We got a response, but it's kinda hard to read when it's all laid out on one line...so...

---

# 3. Let's make the Answer we got from Coingecko more readable

```python
import json

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON and prettify it with indentation
    pretty_response = json.dumps(response.json(), indent=4)  # `indent=4` makes the JSON easier to read
    print(pretty_response)
else:
    # If there was an error, print the status code and reason
    print(f"Error: {response.status_code}, {response.reason}")
```

### Yay! We got the price of BTC.
